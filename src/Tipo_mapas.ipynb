```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
%matplotlib inline
```

```python
base = pd.read_csv('../bases/tipos_mapas_conceituais.csv')
```

```python
sorted(base['Tipo de Mapa'].astype('str').unique())
```

```python
dimensoes = 12, 5
```

```python
plt.figure(figsize=dimensoes)
sns.countplot(x = 'Ano', data=base, hue='Tipo de Mapa')
plt.ylabel('Número de publicações')
```

```python
plt.figure(figsize=dimensoes)
sns.countplot(x = 'Tipo de Mapa', data=base, hue='Ano')
plt.ylabel('Número de publicações')
```
